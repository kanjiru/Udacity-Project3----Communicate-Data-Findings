## Investigation Overview


> In this analysis, The aim is to look at how the duration of trips of customers and subscibers and their locations.The aim is to demonstrate the importance e of data visualization in the data analysis process.


## Dataset Overview

> The selected data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.The data features include duration (secs) and others such as DateTime, customer type, start and end station name, start and end station longitude and latitude, among others.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("201902-fordgobike-tripdata.csv")
```

##  User_types distribution of trips

> More than 80% of trips are taken by Subsribers

```python
# same distribution in countplot format
plt.figure(figsize=(7,7))

sb.set(style='darkgrid')

sb.countplot(data=df, x='user_type',orient= 'h',hue='user_type', dodge=False,
palette="Blues", order=df['user_type'].value_counts().index)
plt.xlabel('User_type', fontsize=14)
plt.ylabel('Count of rides', fontsize=15)
plt.title('Distribution of count of bike rides per user_type', fontsize=17);
```

## Stations with most trips


> Market St at 10th has most trips and the list is Embarcadero at Sansome St

```python
# keeping only top 8 locations with high traffics,
locs = ['San Francisco Ferry Building (Harry Bridges Plaza)','The Embarcadero at Sansome St','San Francisco Caltrain (Townsend St at 4th St)','San Francisco Caltrain Station 2  (Townsend St at 4th St)','Market St at 10th St','Montgomery St BART Station (Market St at 2nd St)','Berry St at 4th St','Powell St BART Station (Market St at 4th St)']
df= df.loc[df['start_station_name'].isin(locs)]

# order from top 1 to top 8
st_stat_counts = df['start_station_name'].value_counts()
st_stat_order = st_stat_counts.index
st_stat_counts = df['start_station_name'].value_counts()
st_stat_order = st_stat_counts.index

base_color = sb.color_palette()[0]
sb.countplot(data=df, y='start_station_name', color=base_color, order=st_stat_order)
plt.xlabel('Count')
plt.ylabel('Start Station Name')
plt.title('Top 8 Trips Stations with the most trips in SF');
```

## Customer type trip duration

> Costomer trips take on average longer as compared to subscribers.


```python
samples = np.random.choice(df.shape[0], 2000, replace = False)
samp = df.loc[samples,:]

plt.figure(figsize = [6, 6])
def boxgrid(x, y, **kwargs):
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color);

g = sb.PairGrid(data = samp, y_vars = 'duration_sec', x_vars =  'user_type',
                size = 7, aspect = 1.5).set(title='Duration of time against user_type');
g.map(boxgrid);
plt.ylim(0,3000);
plt.show()
```
