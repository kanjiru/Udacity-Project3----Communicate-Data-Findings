## Univariate Exploration

> What is the diference in count between user_type (Customer vs Subscriber)?

```python
sb.histplot(x='user_type', data=df);
ax = plt.gca()
ax.set_title("Difference in number between user types")
```

> What is the relationship of diferent non objet/string types in the dataset?

```python
#On the kernel terminal i have updated seaborn to work with later functions e.g histplot
cols = 5
rows = 5
num_cols = df.select_dtypes(exclude='object').columns
fig = plt.figure( figsize=(cols*5, rows*5))
sb.set(rc={'figure.figsize':(11.7,8.27)})
for i, col in enumerate(num_cols):
    
    ax=fig.add_subplot(rows,cols,i+1)
    
    sb.histplot(x = df[col], ax = ax)
    
fig.tight_layout()  
plt.show()

ax = plt.gca()
ax.set_title("Histogram for non string types");
```

> From the histograms above we can deduce that Duration period and distance the distribution is rightly skewed <br>
> The start year depicts data available is of just one year <br>
> From the station_id we can conclude that some stations have a higher number of traffic as compared to others

> What is the diference in density distribution of gender types?

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.kdeplot(x='duration_min', data=df, hue='member_gender',log_scale=True)

ax = plt.gca()
ax.set_title("Difference in Number of gender types")
```

> Men ride bikes more as compared to other and female gender 

> How much distance on average is traveled?

```python
sb.violinplot(x=df['distance_km'],bw=.15,scale='area')

ax = plt.gca()
ax.set_title("Avarege distance travelled")
```

> The avarage distance in kilimeters traveled is about 2KM

> What is the number of bike rides for each day?

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.histplot(x='start_day', data=df)

ax = plt.gca()
ax.set_title("Number of bike rides for each day")
```

> Thursday had the most number of bike rides follwed by teusday <br>
> There is no significant difference in bike rides between Saturday and Sunday one can conclude the bikes are maily used to commute to work.

## Bivariate Exploration

>What is the distance on bike for gender types?

```python
sb.barplot(
    x='member_gender', 
    y='distance_km', 
    data=df, 
   );

ax = plt.gca()
ax.set_title("Distance on bike for gender types")
```

> Other gender travelled the most distance followed by female and lastly males

> What is the duration on the bikes for gender types?

```python
sb.barplot(
    x='member_gender', 
    y='duration_min', 
    data=df, 
   );

ax = plt.gca()
ax.set_title("Duration on bike for gender types")
```

> Other gender travelled rode the bike for a long time followed by females and lastly males

> What is the avarage distance traveled per day?

```python
plt.figure(
    figsize=(12, 6))
sb.stripplot(
    x='start_day', 
    y='distance_km', 
    data=df, 
    linewidth=.5, 
    color='blue')
plt.xticks(rotation=90);

ax = plt.gca()
ax.set_title(" Distance per day")
```

> There is no significant diference of the distance travelled on any day of the week.

> What is the relationship between distance and time on bike?

```python

sb.jointplot(
    x='duration_min', 
    y='distance_km', 
    data=df);

ax = plt.gca()
ax.set_title("Relationship between distance and time on bike")
```

> There is little to no correlation between distance and time spent of the bike.

> What is the distance travelled for each gender type?

```python
sb.displot(df,x="member_gender", y="distance_km")  
sb.set(rc={'figure.figsize':(16.7,11.27)})
ax = plt.gca()
ax.set_title("Distance travelled for genders")
```

> On avarage more men traveled futher in terms of distance

## Multivariate Exploration

> What is the distace travelled in time for each user type?

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.scatterplot(
    x='duration_min',
    y='distance_km', 
    data=df,
    palette='bright',
    hue='user_type');

ax = plt.gca()
ax.set_title("Distace travelled in time for user_type")
```

>customers spend more time on average compare to subscribers and travell further.

> What is the distance travelled in time for each gender?

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.lmplot(
    x="duration_min", 
    y="distance_km", 
    data=df,
    palette='bright',
    hue="member_gender");

ax = plt.gca()
ax.set_title("Distance travelled in time for genders")
```

> Men tend to travel further in respect to time followed my females and finally other gender

> Who shares bikes more often accross user types?

```python
plt.title('Trip bike share between genders')
sb.barplot(
    x='member_gender', 
    y='distance_km', 
    data=df, 
    palette='Paired',
    hue='bike_share_for_all_trip');
```

> Less females tend to share bikes then males and other gender tend to equally share than not.
